A palindromic number reads the same both ways. For example, 9009, 8888, 1221

Find the largest palindrome made from the product of two 3-digit numbers.

```python
def isPalindrome(number):

    numstr = str(number)

    numlength = len(numstr)

    if numlength % 2 == 0:

        numstr_left = numstr[:int(numlength/2)]
        numstr_right = numstr[int(numlength/2):]

        if numstr_left == numstr_right[::-1]:
            return True
        else:
            return False
        
    else:

        numstr_left = numstr[:int(numlength//2)]
        numstr_right = numstr[int((numlength//2)+1):]

        if numstr_left == numstr_right[::-1]:
            return True
        else:
            return False
    
```

```python
max_palindrome = 0

for i in range (1000):
    for j in range(1000):

        prod = i*j 

        if isPalindrome(prod):
            max_palindrome = max(max_palindrome, prod)

print(max_palindrome)
```
