Find the sum of all the primes below two million.

```python
def listPrimes(N):

    primes = [2]

    for i in range(3,N,2):

        primeStatus = True
    
        for prime in primes:
            
            if prime < i//2:
                
                if i % prime == 0:
                    
                    primeStatus = False
                    break

            else:

                break

        if primeStatus:

            primes.append(i)

    return primes

print(listPrimes(100))

        
```
