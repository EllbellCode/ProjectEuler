If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are $3 + 3 + 5 + 4 + 4 = 19$ letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

**NOTE:** Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
single = {
    "0": "", "1": "one", "2": "two", "3": "three", "4": "four", "5": "five",
    "6": "six", "7": "seven", "8": "eight", "9": "nine"
}

teens = {
    "1": "eleven", "2": "twelve", "3": "thirteen", "4": "fourteen", "5": "fifteen",
    "6": "sixteen", "7": "seventeen", "8": "eighteen", "9": "nineteen"
}

tens = {
    "1": "ten", "2": "twenty", "3": "thirty", "4": "forty", "5": "fifty",
    "6": "sixty", "7": "seventy", "8": "eighty", "9": "ninety"
}

hundreds = {
    "1": "onehundred", "2": "twohundred", "3": "threehundred", 
    "4": "fourhundred", "5": "fivehundred", "6": "sixhundred", 
    "7": "sevenhundred", "8": "eighthundred", "9": "ninehundred"
}

thousand = "one thousand"
```

```python
stringSum = 0

for i in range(1,1000):

    numstr = str(i)
    length = len(numstr)

    if length == 1:

        string = single[numstr[0]]

    elif length == 2:

        if numstr[0] == '1' and numstr[1] != '0':

            string = teens[numstr[-1]]

        else:

            string = tens[numstr[0]] + single[numstr[1]]

    elif length == 3:

        string = hundreds[numstr[0]]

        numstr = numstr[1:]

        if numstr[0] == '1' and numstr[1] != '0':

            string += "and" + teens[numstr[-1]]

        elif numstr[0] != '0':

            string +=  "and" + tens[numstr[0]] + single[numstr[1]]

        elif numstr[0] == '0' and numstr[1] != '0':

            string += "and" + single[numstr[1]]


    stringSum += len(string)

    print(f"Number: {string}, length: {len(string)}")

stringSum += len("onethousand")

print(stringSum)

    

```
