By listing the first six prime numbers: 2,3,5,7,11,13, we can see that the 6th prime is 13.

What is the  10001st prime number?

```python
def isPrime(number):

    i = 2
    prime = True
    while i < number//2:

        if number % i == 0:

            prime = False
            break

        else:

            i += 1

    return prime

print(isPrime(25))

```

```python
primes = [2]
number = 3

while len(primes) < 10001:

    if isPrime(number):

        # print(f"{number} is Prime")
        primes.append(number)

    number += 2

print(primes[-1])


```
