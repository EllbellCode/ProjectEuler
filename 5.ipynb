2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible (divisible with no remainder) by all of the numbers from 1 to 20 ?

```python
# Approach is grounded in the fact that the Lowest Common Multiple (LCM) of a given set of numbers is equal to the product of the prime factors of the highest powers from all of the numbers
# For example, the LCM of 2,4,8,3 is 34 because the highest power of 2 is 3 (8) and the highest power of 3 is 1 so 2**3 * 3 = 24

divisors = [x for x in range(1,21)]
```

```python
primes = [2,3,5,7,11,13,17,19]

def primeFactor(number):

    factors = []

    for prime in primes:

        if number % prime == 0:

            scratch = number

            while scratch % prime == 0:
                
                factors.append(prime)
                scratch = scratch/prime
        
    return factors

print(primeFactor(20))
```

```python
from collections import defaultdict

# if you query a key in defaultdict that does not exist, it automatically creates that key with value 0
ocurrences = defaultdict(int)

for divisor in divisors:

    factors = primeFactor(divisor)

    for factor in factors:

        count = factors.count(factor)

        # ocurrences[factor] becomes 0 if the factor is not yet a key. Thanks to default dict!
        ocurrences[factor] = max(ocurrences[factor], count)

print(ocurrences)

total = 1

for prime in ocurrences.keys():

    total *= prime**ocurrences[prime]

print(total)
```
