What is the largest prime factor of 600851475143?

```python
# iterate through numbers (while number < 0.5*N)
# check if number is prime. If so, see if N % number == 0
# If so, divide the number and set it to the new number
# Store the number we just divided by
from itertools import chain

maxNum = 0
N = 600851475143
number = 2

while number < N or number == N:

    prime = True
    if number % 1000 == 0:
        print(number)
    # print(f"Testing Primality of {number}...")
    for i in chain([2], range(3, (number//2)+1, 2)):

        if number % i == 0:
            prime = False
            # print(f"{number} is not Prime: {i} divides")
        

    if prime:
        # print(f"{number} is prime!")
        if N % number == 0:
            print(f"{number} is prime and divides {N}!")
            N = N/number
            print(f"New Number: {N}")
            maxNum = number
        else:
            # print(f"{number} does not divide {N}")
            number += 1

    else:
        number +=1

print(f"Largest Prime Factor: {maxNum}")



```
